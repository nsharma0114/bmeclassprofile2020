```python
df.columns = np.arange(len(df.columns))
```

```python
def getdata(df):
    data_all = df
    data = []
    if type(data_all[0]) == str:
        for i,d in enumerate(data_all): 
            data.extend(d.split(',')) 
        data = pd.Series(data)     
    else:
        data = data_all 
    return data 
```

```python
def plotdata(data, n, title, xlab, ylab): 
    datay = data.value_counts().values[0:n]
    datax = data.value_counts().index[0:n]
    plt.figure(figsize=(10,6))
    sns.set(style="whitegrid") 
    ax = sns.barplot(x=datax, y=datay) 
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
    ax.set_title(title)
    ax.set_xlabel(xlab)
    ax.set_ylabel(ylab)
```

```python
TEs = getdata(df[18].values)
```

```python
title = "Most Popular Technical Electives"
xlab = "Course Code"
ylab = "# of BME Students Enrolled"
plotdata(TEs,36,title,xlab,ylab)
```
